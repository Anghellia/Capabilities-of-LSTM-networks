```python
# Импортируем библиотеки

%tensorflow_version 1.x
!pip install pyinform
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error
from pyinform.dist import Dist
from pyinform.shannon import entropy
```

```python
def split_sequence(sequence, window):
    n = len(sequence)
    X, y = list(), list()
    for i in range(n):
        end_ix = i + window
        if end_ix > n-1:
            break
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)


def predict_next_value(sequence):
    sequence = sequence[:window]
    sequence = sequence.reshape((1, window, 1))
    prediction = model.predict(sequence, verbose=0)[0][0]
    return np.round(prediction)


def test_performance(sequence, window):
    sequence = sequence * 2
    X_test, y_test = split_sequence(sequence, window)
    for i in zip(X_test, y_test):
        prediction = predict_next_value(i[0])
        print(f'Expected: {i[1]}, Predicted: {prediction}')
```

```python
def calcualte_entropy(sequence):
    d = Dist(max(sequence) + 1)
    for event in sequence:
        d.tick(event)
    return entropy(d)
```

```python
def generate_sequence():

    X, y = list(), list()

    for i in range(1, 10):
        
        array = [i, 0, 0, 0] * 2
        n = len(array)

        for j in range(n):
            end_ix = j + 4
            if end_ix > n - 1: break
            seq_x, seq_y = array[j:end_ix], array[end_ix]
            X.append(seq_x)
            y.append(seq_y)

    X, y = np.array(X), np.array(y)
    X = X.reshape((X.shape[0], X.shape[1], 1))
    
    return X, y
```
